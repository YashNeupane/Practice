```python
# List declaration

fruits = ["apple", "orange", "strawberry"]
print(fruits[1])
fruits[1] = "watermelon"
print(fruits[1])

listExample = [1, 2, ["a", "b", [True, False]]]
print(listExample[0])
print(listExample[2][2][1])
# print(listExample[2][3])  # the index 3 exceeds the maximum index in listExample[2]
print(listExample[2][1])
print(listExample[-1])
print(listExample[-1][-2][0])
print(listExample[-1][-2][-1])
print(listExample[-1][0])
```

```python
# mutators for list
classSchedule = ["introduction", "variables", "list"]

# 1. we want to change classSchedule to ["introduction", "variables", "list", "selection"]
classSchedule.append("selection")
# equivalent to:
# classSchedule.insert(3, "selection")
print(classSchedule)
# list.append(x) adds x to the last position of list
# append() does not return any value, so you should not use assignment operator with it.
# classSchedule = classSchedule.append("selection")
# print(classSchedule)

#  2. we want to change the schedule to ["introduction", "variables", "list", "loop", "selection"]
classSchedule.insert(3,"loop")
print(classSchedule)
# list.insert(i, x) inserts x to the position with index i in list
# insert() does not return any value, so you should not use it with the assignment operator

# 3. we want to add all topics in controlStructures to the end of classSchedule
classSchedule = ["introduction", "variables", "list"]
controlStructures = ["selection", "loop", "method"]

classSchedule.extend(controlStructures)
print(classSchedule)
# list1.extend(list2) appends all elements in list2 to list1
# extend() does not return any value, so you should not use it with the assignment operator

# the difference between extend and append
list1 = [1,2,3]
list2 = [4,5,6]
list3 = [1,2,3]
list4 = [4,5,6]
list1.append(list2)
list3.extend(list4)
print(list1)
len(list1)
print(list3)
len(list3)
```

```python
# 4. we want to remove "method" from the class schedule
classSchedule.remove("method")
print(classSchedule)
# ValueError: list.remove(x): x not in list
# list.remove(x) removes the first occurence of x in list
# remove() does not return any value

# 5. we want to remove "loop" from the class schedule
# classSchedule.remove("loop")
# equivalent to
x = classSchedule.pop()
print(classSchedule)
print(x)
# list.pop() removes the last element in a list
# pop() returns the value that is removed: x = classSchedule.pop()
# list.pop(i) removes the element indexed by i in a list
classSchedule.pop(1)
print(classSchedule)

# suppose I want to remove the value "list" whose index is 1
# classSchedule.remove("list")
# classSchedule.pop(1)
# pop() is more efficient because remove() requires python to search from the 1st element afterward
# until it finds "list" in a list. pop() specifies the index, so python could directly go to
# that position and  the search process is skipped. Thus, if you can specify the index when writing
# the program, do it and use pop() instead of remove().
```

```python


# 6. we want to reverse the element sequence
classSchedule.reverse()
print(classSchedule)
# list.reverser() reverse the element sequence in list
# reverse() method does not return anything

# 7. we want to sort elements in a list
classSchedule.sort()
print(classSchedule)
# list.sort() sorts elements in a list in the ascending order
classSchedule.sort(reverse=True)
print(classSchedule)
# by providing the option reverse=True, sort() sorts elements in the descending order

listExample = [1, 2, "Python", True]
listExample.sort()  # this will lead to an error because we cannot compare int and str in python
print(listExample)

listExample = [1, 2, False]
listExample.sort()
print(listExample)
```
